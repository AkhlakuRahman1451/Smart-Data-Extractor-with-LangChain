```python
from langchain_groq import ChatGroq
chatbot = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
from langchain_core.messages import HumanMessage

messageToTheChatbot = [
    HumanMessage(content="My favorite color is blue. What is your favorite color?")
]
```

## Call the ChatModel(the LLM)

```python
chatbot.invoke(messageToTheChatbot)
```

## add memory our chatbot

```python

from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import FileChatMessageHistory
```

```python

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI chatbot."),
    MessagesPlaceholder(variable_name="chat_history"), 
    ("human", "{input}")                              
])
```

```python

chain = prompt | chatbot
```

```python

def get_chat_history(session_id: str):
    return FileChatMessageHistory(f"chat_history_{session_id}.json")
```

```python

conversational_chain = RunnableWithMessageHistory(
    chain,
    get_chat_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)
```

```python

conversational_chain = RunnableWithMessageHistory(
    chain,
    get_chat_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)
print("✅ মেমোরিসহ চ্যাটবট চেইন সম্পূর্ণ রেডি!")
```

```python

response1 = conversational_chain.invoke(
    {"input": "Hello! My name is Akhlaq. I am  AI engineer."},
    config={"configurable": {"session_id": "user_1"}}
)
print("Bot:", response1.content)
print("-" * 50)
```

```python

response2 = conversational_chain.invoke(
    {"input": "What is my name and profession?"},
    config={"configurable": {"session_id": "user_1"}}
)
print("Bot:", response2.content)
```
